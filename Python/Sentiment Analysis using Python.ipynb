```python
class analysis_text:
    def center(self, toplevel):
        toplevel.update_idletasks()
        w = toplevel.winfo_screenwidth()
        h = toplevel.winfo_screenheight()
        size = tuple(int(_) for _ in toplevel.geometry().split('+')[0].split('x'))
        x = w/2 - size[0]/2
        y = h/2 - size[1]/2
        toplevel.geometry("%dx%d+%d+%d" % (size[0], size[1], x, y))

    def callback(self):
        if messagebox.askokcancel("Quit", "Do you want to leave?"):
            self.main.destroy()

    def setResult(self, type, res):
        if type == "neg":
            self.negativeLabel.configure(text="Negative sentiment: " + str(res) + "%")
        elif type == "neu":
            self.neutralLabel.configure(text="Neutral sentiment: " + str(res) + "%")
        elif type == "pos":
            self.positiveLabel.configure(text="Positive sentiment: " + str(res) + "%")

    def runAnalysis(self):
        sentence = self.line.get()
        sid = SentimentIntensityAnalyzer()
        ss = sid.polarity_scores(sentence)
        if ss['compound'] >= 0.05:
            self.normalLabel.configure(text="Sentiment Analysis Result: Positive feeling", fg="green")
        elif ss['compound'] <= -0.05:
            self.normalLabel.configure(text="Sentiment Analysis Result: Negative feeling", fg="red")
        else:
            self.normalLabel.configure(text="Sentiment Analysis Result: Neutral feeling")
        for k in ss:
            self.setResult(k, ss[k] * 100)

    def editedText(self, event):
        self.typedText.configure(text=self.line.get() + event.char)

    def runByEnter(self, event):
        self.runAnalysis()

    def __init__(self):
        self.main = Tk()
        self.main.title("Sentiment Analysis System")
        self.main.geometry("600x400")
        self.main.resizable(width=False, height=False)
        self.main.protocol("WM_DELETE_WINDOW", self.callback)
        self.center(self.main)

        Label(self.main, text="Please type your comment:").pack()
        self.line = Entry(self.main, width=70)
        self.line.pack()
        self.typedText = Label(self.main, text="", fg="blue", font=("Helvetica", 20))
        self.typedText.pack()
        self.line.bind("<Key>", self.editedText)
        self.line.bind("<Return>", self.runByEnter)

        self.negativeLabel = Label(self.main, text="", fg="red", font=("Times New Roman", 20))
        self.negativeLabel.pack()
        self.neutralLabel = Label(self.main, text="", font=("Times New Roman", 20))
        self.neutralLabel.pack()
        self.positiveLabel = Label(self.main, text="", fg="green", font=("Times New Roman", 20))
        self.positiveLabel.pack()
        self.normalLabel = Label(self.main, text="", font=("Times New Roman", 20))
        self.normalLabel.pack()


myanalysis = analysis_text()
mainloop()
```
